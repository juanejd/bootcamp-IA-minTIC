## limpiar los datos eliminando las filas con valores nulos


```python
inventory = inventory.dropna()
satisfaction = satisfaction.dropna()
sales = sales.dropna()
```

## 4.


### Calcular ventas totales por tienda y producto


```python
total_sales_store = sales.groupby("ID_Tienda")["Cantidad_Vendida"].sum().reset_index()
total_sales_store
```

```python
total_sales_product = sales.groupby("Producto")["Cantidad_Vendida"].sum().reset_index()
total_sales_product
```

### Calcula los ingresos totales por cada producto en la tienda


```python
sales["ventas_totales"] = sales["Cantidad_Vendida"] * sales["Precio_Unitario"]
sales.head()
```

```python
sales.describe()
```

### calcula el promedio de ventas por tienda


```python
mean_sales_store = sales.groupby("ID_Tienda")["ventas_totales"].mean().reset_index()
mean_sales_store
```

### Calcula el promedio de ventas por producto


```python
mean_sales_product = sales.groupby("Producto")["ventas_totales"].mean().reset_index()
mean_sales_product
```

## 5.


```python
# merge ---> une dos tablas. on = columna que se relaciona
inventory_test = pd.merge(sales, inventory, on="ID_Tienda")
inventory_test
```

```python
# relacionamos el ID_Tienda y el Producto de las tablas sales y inventory y hacemos un inner join
sales_inventory = pd.merge(sales, inventory, on=["ID_Tienda", "Producto"], how="inner")
sales_inventory
```

### rotacion de inventario


```python
rotation_inventory = sales_inventory.copy()  # copiamos la tabla sales_inventory
rotation_inventory["rotacion_inventario"] = (
    rotation_inventory["Cantidad_Vendida"] / rotation_inventory["Stock_Disponible"]
)
rotation_inventory
```

### productos con ventas menores al 10% del stock disponible


```python
# se crea un dataframe con data de sales_inventory y solo tomamos las columnas ID_Tienda y Producto
products_critical = pd.DataFrame(sales_inventory, columns=["ID_Tienda", "Producto"])

# se calcula el porcentaje de ventas y el porcentaje de stock disponible
products_critical["percentage_sales"] = sales_inventory["Cantidad_Vendida"] / 100
products_critical["percentage_stock"] = (
    sales_inventory["Stock_Disponible"] * 0.10
) / 100

# se crea una columna con resultado booleanos en caso de que se cumpla la condicion
products_critical["inventory_critical"] = (
    products_critical["percentage_sales"] < products_critical["percentage_stock"]
)
products_critical
```

### tiendas con inventario critico


```python
inventory_store_critical = (
    products_critical.groupby("ID_Tienda")["inventory_critical"].sum().reset_index()
)
inventory_store_critical
```

### Productos con inventario critico


```python
inventory_products_critical = (
    products_critical.groupby("Producto")["inventory_critical"].sum().reset_index()
)
inventory_products_critical
```

```python
print(
    f"Hay {inventory_store_critical['inventory_critical'].sum()} tiendas con ventas menores al 10% del stock disponible"
)
```

### agrupamos por tienda, ventas y satisfaccion


```python
satisfaction_sales_analysis = (
    satisfaction_stores.groupby("ID_Tienda")[
        ["ventas_totales", "Satisfacción_Promedio"]
    ]
    .mean()
    .reset_index()
)
satisfaction_sales_analysis
```
